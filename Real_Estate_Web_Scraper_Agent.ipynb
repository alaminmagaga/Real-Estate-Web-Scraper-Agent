```python
from crewai import Agent, Task, Crew
from crewai_tools import SerperDevTool, ScrapeWebsiteTool
import os

os.environ["SERPER_API_KEY"] = "your api key"
```

```python
# Web Scraper Agent
web_scraper_agent = Agent(
    role='Real Estate Web Scraper Agent',
    goal="""Scrape housing data, including property titles, prices, and links, from the specified real estate website.""",
    backstory="""You are an AI agent specialized in scraping real estate housing data from specified websites.""",
    verbose=True
)

# Tool for web scraping
scrape_tool = ScrapeWebsiteTool(website_url="https://nigeriapropertycentre.com/for-sale")

# Task: Scrape real estate housing data
scrape_housing_task = Task(
    description="Scrape housing data, including property titles, prices, and links from the specified real estate website.",
    expected_output="""A list of properties with:
    - Title
    - Price
    - Link to the property""",
    agent=web_scraper_agent,
    tools=[scrape_tool],
    output_file='data.md'
)

# Create a crew with the agent and task
crew = Crew(
    agents=[web_scraper_agent],
    tasks=[scrape_housing_task],
    verbose=True
)

# Function to execute the web scraping task
def scrape_housing_data():
    result = crew.kickoff()
    print("\nScraped Housing Data:\n")
    print(result)

# Run the program
if __name__ == "__main__":
    scrape_housing_data()
```
