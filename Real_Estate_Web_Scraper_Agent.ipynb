```python
from crewai import Agent, Task, Crew
from crewai_tools import SerperDevTool, ScrapeWebsiteTool, FileWriterTool
import csv
import os

# Web Scraper Agent
web_scraper_agent = Agent(
    role='Real Estate Web Scraper Agent',
    goal="""Scrape housing data, including property titles, prices, and links, from the specified real estate website.""",
    backstory="""You are an AI agent specialized in scraping real estate housing data from specified websites.""",
    verbose=True
)

# Tool for web scraping
scrape_tool = ScrapeWebsiteTool(website_url="https://nigeriapropertycentre.com/for-sale?page=1")

# Task: Scrape real estate housing data
scrape_housing_task = Task(
    description="Scrape housing data, including property titles, prices, and links from the specified real estate website.",
    expected_output="""A list of properties with:
    - Title
    - Price
    - Link to the property""",
    agent=web_scraper_agent,
    tools=[scrape_tool]
)

# Create a crew with the agent and task
crew = Crew(
    agents=[web_scraper_agent],
    tasks=[scrape_housing_task],
    verbose=True
)

# Initialize FileWriterTool
file_writer_tool = FileWriterTool()

# Function to execute the web scraping task and write to a CSV file
def scrape_housing_data_and_write_csv():
    # Execute the scraping task
    scraped_data = crew.kickoff()

    # Process the scraped data
    # Assuming the scraped data is in a list of dictionaries
    housing_data = scraped_data.get('output', [])
    if not housing_data:
        print("No data scraped.")
        return

    # CSV file details
    filename = "housing_data.csv"
    directory = "scraped_data"
    filepath = os.path.join(directory, filename)

    # Create CSV content
    headers = ['Title', 'Price', 'Link']
    csv_content = [headers]
    for entry in housing_data:
        csv_content.append([entry.get('title'), entry.get('price'), entry.get('link')])

    # Convert CSV content to a string format
    csv_string = "\n".join([",".join(map(str, row)) for row in csv_content])

    # Write CSV content to the file
    result = file_writer_tool._run(filename=filename, content=csv_string, directory=directory)
    print("\nFile Write Result:\n", result)

# Run the program
if __name__ == "__main__":
    scrape_housing_data_and_write_csv()
```

```python
from crewai import Agent, Task, Crew
from crewai_tools import SerperDevTool, ScrapeWebsiteTool

# Web Scraper Agent
web_scraper_agent = Agent(
    role='Real Estate Web Scraper Agent',
    goal="""Scrape housing data, including property titles, prices, and links, from the specified real estate website.""",
    backstory="""You are an AI agent specialized in scraping real estate housing data from specified websites.""",
    verbose=True
)

# Function to create a ScrapeWebsiteTool for a given page
def create_scrape_tool(page_number):
    return ScrapeWebsiteTool(website_url=f"https://nigeriapropertycentre.com/for-sale?page={page_number}")

# Multi-Page Scraper Agent
multi_page_scraper_agent = Agent(
    role='Multi-Page Real Estate Scraper Agent',
    goal="""Scrape housing data across multiple pages (1 to 10) from the specified real estate website.""",
    backstory="""You are an AI agent tasked with scraping real estate housing data from multiple pages of the specified website.""",
    verbose=True
)

# Task: Scrape real estate housing data across multiple pages
multi_page_scraper_task = Task(
    description="Scrape housing data, including property titles, prices, and links, from pages 1 to 10 of the specified real estate website.",
    expected_output="""A compiled list of properties with:
    - Title
    - Price
    - Link to the property""",
    agent=multi_page_scraper_agent
)

# Function to execute the multi-page scraping task
def scrape_housing_data():
    all_results = []
    for page_number in range(1, 11):  # Loop through pages 1 to 10
        print(f"Scraping page {page_number}...")
        scrape_tool = create_scrape_tool(page_number)
        crew = Crew(
            agents=[multi_page_scraper_agent],
            tasks=[Task(
                description=f"Scrape housing data from page {page_number}.",
                expected_output="""A list of properties with:
                - Title
                - Price
                - Link to the property""",
                agent=multi_page_scraper_agent,
                tools=[scrape_tool]
            )],
            verbose=True
        )
        result = crew.kickoff()
        all_results.append(result)
    print("\nScraped Housing Data:\n")
    for res in all_results:
        print(res)

# Run the program
if __name__ == "__main__":
    scrape_housing_data()
```

```python
from crewai import Agent, Task, Crew, Process
from crewai_tools import ScrapeWebsiteTool

# Agent 1: URL Listing Agent
url_lister_agent = Agent(
    role='URL Listing Agent',
    goal="Generate a list of URLs for pages 1 to 5 of the specified real estate website.",
    verbose=True,
    memory=True,
    backstory="You are an AI agent tasked with listing URLs for subsequent web scraping tasks.",
)

# Agent 2: Web Scraper Agent
web_scraper_agent = Agent(
    role='Real Estate Web Scraper Agent',
    goal="Scrape housing data, including property titles, prices, and links, from the specified real estate website.",
    verbose=True,
    memory=True,
    backstory="You are an AI agent specialized in scraping real estate housing data from specified websites.",
)

# Task 1: Generate URLs
url_listing_task = Task(
    description="Generate a list of URLs for pages 1 to 5 of the real estate website.",
    expected_output="A list of URLs in the format:\n"
                    "https://nigeriapropertycentre.com/for-sale?page=1\n"
                    "https://nigeriapropertycentre.com/for-sale?page=2\n"
                    "... up to page 5.",
    agent=url_lister_agent,
    tools=[],
)

# Task 2: Scrape housing data
scrape_tasks = [
    Task(
        description="Scrape housing data from listed website urls",
        expected_output="A list of properties with:\n"
                        "- Title\n"
                        "- Price\n"
                        "- Link to the property.",
        agent=web_scraper_agent,
        tools=[ScrapeWebsiteTool],
    )
    for i in range(1, 6)
]

# Workflow: Single crew handling both agents and tasks
def scrape_housing_data():
    # Define a single crew for both tasks
    crew = Crew(
        agents=[url_lister_agent, web_scraper_agent],
        tasks=[url_listing_task] + scrape_tasks,  # Combine URL listing and scraping tasks
        process=Process.sequential,  # Sequential task execution
        verbose=True
    )

    # Execute all tasks sequentially
    results = crew.kickoff()

    # Print the results
    print("\nResults:")
    for output in results["outputs"]:
        print(output)

# Run the program
if __name__ == "__main__":
    scrape_housing_data()
```
